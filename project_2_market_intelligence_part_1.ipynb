⚙️ Technical Note: Handling Different CSV Separators
In Python's Pandas library, the read_csv() function defaults to a comma (,) as a separator. However, depending on the region (like in Brazil) or how the file was exported (e.g., from Excel), the separator might be a semicolon (;).

* Standard: pd.read_csv(path) (assumes comma)

* Custom: pd.read_csv(path, sep=';') (for semicolon-separated files)

**Why it matters:** If the wrong separator is used, Pandas will load the entire row as a single column, causing errors during data cleaning or database migration.

#### 2.5.3) 🚀 Final Data Cleansing & Migration

In this section, I migrate the `listings` table. This table acts as a bridge between properties, their addresses, and their owners. To ensure database integrity and avoid Foreign Key errors, I normalize the IDs by converting them to strings, ensuring they match the VARCHAR format in the MySQL database.

**Strategy:** To fix **Error 1452**, I used a Filtering Strategy. I compared the listings DataFrame against the existing IDs in the owners and addresses tables. By using the .isin() method, I ensured that only 'valid' records were sent to the database, preventing the migration from crashing and maintaining 100% data integrity.

```python
# --- Table 4: LISTINGS ---

# 1. Path to the file
path_listings = r"C:\Users\vanes\OneDrive\Área de Trabalho\PROJETOS EM ANDAMENTO\PORTFÓLIO\PROJETO 2\Parte_2\csv\4-hospedagens.csv"

# 2. Load the CSV
df_listings = pd.read_csv(path_listings, encoding='utf-8')

# 3. Renaming columns to English (Prevents KeyError)
df_listings.columns = ['listing_id', 'property_type', 'address_id', 'owner_id', 'is_active']


# 4. Normalize IDs to String (to match VARCHAR in MySQL)
for col in ['listing_id', 'address_id', 'owner_id']:
    df_listings[col] = df_listings[col].astype(str).str.replace('.0', '', regex=False)

# 5. DIAGNOSIS & CORRECTION
# Get valid IDs from the parent tables already in MySQL
db_owners = pd.read_sql("SELECT owner_id FROM owners", con=engine)['owner_id'].tolist()
db_addresses = pd.read_sql("SELECT address_id FROM addresses", con=engine)['address_id'].tolist()

# 5.1. Filter the listings: Only keep rows where parent IDs exist in the database
df_listings_clean = df_listings[
    df_listings['owner_id'].isin(db_owners) & 
    df_listings['address_id'].isin(db_addresses)
].copy()

# 5.2. Calculate how many rows were removed for the report
removed_count = len(df_listings) - len(df_listings_clean)

# 6. Drop Duplicates
df_listings_clean = df_listings_clean.drop_duplicates(subset=['listing_id'])

# 7. Migration
# try:
#     df_listings_clean.to_sql('listings', con=engine, if_exists='append', index=False)
#     print(f"✅ Success! {len(df_listings_clean)} listings migrated.")
#     print(f"⚠️ Note: {removed_count} orphaned rows were excluded to maintain integrity.")
# except Exception as e:
#     print(f"❌ Error: {e}")

```

#### 2.5.4) 🔏 Data Integrity Report: 
During the migration of the listings table, I identified 16 orphaned records (0.16% of the dataset) that referenced non-existent owners or addresses. #### 2.6.2) Data Integrity Check - Part 2: Orphaned Records Discovery 

This step cross-references the Bookings dataset against existing Customers and Listings in the database. Initial results show a high volume of mismatched IDs. I am now investigating if this is a genuine data gap or a data type mismatch (Integer vs. String) before proceeding with the final filtering

```python
# 1. Checking if all customers IDs in the Bookings CSV exist in the Customers table.
db_customers = pd.read_sql("SELECT customer_id FROM customers", con=engine)
missing_customers = df_bookings[~df_bookings['customer_id'].isin(db_customers['customer_id'])]



# 2. Checking if all listings IDs in the Bookings CSV exist in the Listings table.
db_listings = pd.read_sql("SELECT listing_id FROM listings", con=engine)
missing_listings = df_bookings[~df_bookings["listing_id"].isin(db_listings["listing_id"])]



print("Missing Customers IDs in database: {}".format(missing_customers['customer_id'].unique()))
print("Missing Listings IDs in database: {}".format(missing_listings["listing_id"].unique()))
```

#### 2.6.3) Data Cleansing - Part 1 - Solving Type Mismatch & Orphaned Records"

Initially, the filtering returned zero records because the IDs in the CSV were being interpreted as integers, while the database IDs were strings.<br> 
After converting the IDs to strings using .astype(str), the integrity check worked correctly."

* Original Records: 10,000

* Records After Cleansing: 9,983

* Findings: Only 17 orphaned records (0.17%) were identified and removed to maintain referential integrity.

```python
# Cleansing Orphaned Records 
df_bookings_cleaned = df_bookings[df_bookings['listing_id'].isin(db_listings['listing_id'])]
df_bookings_cleaned = df_bookings_cleaned[df_bookings_cleaned['customer_id'].isin(db_customers['customer_id'])]

print("Original Records:{}".format(len(df_bookings)))
print("Records after cleansing: {}".format(len(df_bookings_cleaned)))
```

```python
# Ensuring data type consistency by converting IDs to strings for comparison
df_bookings_cleaned = df_bookings[df_bookings['listing_id'].astype(str).isin(db_listings['listing_id'].astype(str))]

df_bookings_cleaned = df_bookings_cleaned[df_bookings_cleaned['customer_id'].astype(str).isin(db_customers['customer_id'].astype(str))]

print("The Bookings database has {} rows.".format(len(df_bookings_cleaned)))
```

#### 2.6.4) Data Cleansing - Part 2 - Drop the duplicates
I implemented a custom validation script to check for duplicate booking_id entries. <br>
This ensures the dataset contains only unique transactions before the final load.

```python
#Save data total before remove duplicates
# Salva o total antes de remover duplicatas
total_before = len(df_bookings_cleaned)

#Remove duplicates
# Remove as duplicatas
df_bookings_cleaned = df_bookings_cleaned.drop_duplicates(subset=["booking_id"])

#Checking if something has changed
# Verifica se algo mudou
total_after = len(df_bookings_cleaned)

if total_before > total_after:
    print(f"Heads up! {total_before - total_after} duplicate bookings were removed.")
else:
    print("No duplicates found. The data is unique!")
```
